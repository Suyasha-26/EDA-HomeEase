```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Lol.csv")
data
```

```python
data.drop(columns = ["Unnamed: 0","Location","Property Title"],inplace=True)
```

```python
data.head()
```

```python
import pandas as pd

# Assuming your DataFrame is named df

# Create a new column 'Affordability' based on the 'Price'
data['Affordability'] = pd.cut(
    data['Price'],
    bins=[0, 2.5e7, 7e7, float('inf')],
    labels=['Highly Affordable', 'Moderately Affordable', 'Luxury']
)

# Display the updated DataFrame
data
```

```python
# Function to convert price into lakhs
def convert_to_lakh(price):
    if pd.isna(price):
        return None
    return f"{price / 1_00_000:.1f} Lakh"

# Apply conversion to your dataframe

data["Price_in_Lakh"] = data["Price"].apply(convert_to_lakh)
```

```python
data
```

```python
# Convert Price (₹) to Lakhs as integer
data["Price_in_Lakh"] = (data["Price"] / 1_00_000).astype(int)
data
```

```python
data["Area_SqFt_Number"] = data["Area_SqFt_Number"].astype(int)
data["Area_per_SqFt_Price"] = data["Area_per_SqFt_Price"].astype(int)
```

```python
data.drop(columns='Price',inplace=True)
```

```python
def plot_numerical_analysis(df , column):
    
    plt.figure(figsize = (15,10))

    #Histogram
    plt.subplot(2,2,1)
    ax = sns.histplot(x = column , data = df , color = 'skyblue' , edgecolor = 'black' , bins = 6,kde=True,line_kws={"color":"Red"})
    plt.title(f'{column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    #KDE plot
    plt.subplot(2,2,2)
    sns.kdeplot(x = column , data = df , color = 'darkred')
    plt.title(f'{column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    #Boxplot
    plt.subplot(2,2,3)
    sns.boxplot(x = column , data = df , color = 'lightgreen')
    plt.title(f'{column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    for container in ax.containers:
        ax.bar_label(container,fontsize = 13,padding=3)

    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# 🧩 UNIVARIATE ANALYSIS: CATEGORICAL COLUMNS
# ==============================

# Identify categorical columns
categorical_cols = data.select_dtypes(include='object').columns

print("📋 Categorical Columns Detected:")
print(list(categorical_cols))
print("="*60)

# Loop through all categorical columns
for col in categorical_cols:
    print(f"\n🔹 COLUMN: {col}")
    print("-" * 60)
    
    # Frequency and percentage distribution
    counts = data[col].value_counts()
    percent = round((data[col].value_counts(normalize=True) * 100), 2)
    summary = pd.DataFrame({'Count': counts, 'Percentage': percent})
    print(summary.head(10))  # show top 10
    
    # -----------------------------
    # 🔸 Bar Chart
    # -----------------------------
    plt.figure(figsize=(8, 4))
    sns.barplot(x=counts.head(10).index, y=counts.head(10).values, palette="pastel", edgecolor="black")
    plt.title(f"Univariate Analysis of {col}", fontsize=13)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    
    # -----------------------------
    # 🔸 Pie Chart
    # -----------------------------
    if len(counts) <= 10:  # only for small category sets
        plt.figure(figsize=(6, 6))
        plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
        plt.title(f"{col} Distribution", fontsize=13)
        plt.tight_layout()
        plt.show()

# ==============================
# 🧠 BONUS INSIGHT: Cross Analysis
# ==============================
if 'City' in data.columns and 'Property Type' in data.columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, x='City', hue='Property Type', palette='pastel', edgecolor='black')
    plt.title("Property Type Distribution by City", fontsize=13)
    plt.xlabel("City")
    plt.ylabel("Count")
    plt.legend(title="Property Type")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
