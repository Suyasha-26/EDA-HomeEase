```python
new_data["Area_SqFt"] = new_data["Property Title"].str.extract(r"(\d{3,5}\s*Sq\.Ft\.)")

# ✅ Extract Property Title (text after 'in' or 'In')
new_data["Extracted_Property_Title"] = new_data["Property Title"].str.extract(r"in\s+(.+)", flags=re.IGNORECASE)
```

```python
new_data.dropna(inplace=True)
```

```python
new_data = new_data.drop(new_data[new_data['Property Type'] == 'Shop'].index)
new_data = new_data.drop(new_data[new_data['Property Type'] == 'Land'].index)
new_data = new_data.drop(new_data[new_data['Property Type'] == 'Warehouse'].index)
new_data = new_data.drop(new_data[new_data['Property Type'] == 'Office Space in IT/SEZ'].index)
```

```python
new_data
```

```python

# Extract only the number part before 'BHK'
new_data["BHK"] = new_data["Unit Type"].str.extract(r"(\d+)(?=\s*BHK)", expand=False)
```

```python
new_data
```

```python
new_data.reset_index(drop=True,inplace=True)
```

```python
new_data
```

```python
new_data.rename(columns = {"Price (₹)":"Price","Area":"Locality"},inplace=True)
```

```python
#new_data['Area_per_SqFt_Price'] = new_data["Price"]/new_data[
# Extract only the digits (area number)
new_data["Area_SqFt_Number"] = new_data["Area_SqFt"].str.extract(r"(\d+)", expand=False)

# Convert to numeric
new_data["Area_SqFt_Number"] = pd.to_numeric(new_data["Area_SqFt_Number"])
```

```python
new_data
```

```python
new_data['Area_per_SqFt_Price'] = new_data["Price"]/new_data['Area_SqFt_Number']
new_data['Area_per_SqFt_Price'] = new_data['Area_per_SqFt_Price'].round(0)
```

```python
new_data
```

```python
new_data.drop(columns=["index","Unit Type","Agent Name","Property Link","Area_SqFt","Extracted_Property_Title"],inplace=True)
```
