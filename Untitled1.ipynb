```python
# Univariate Analysis of Numerical Column
def plot_numerical_analysis(df, column):
    
    plt.figure(figsize = (15,10))

    #Histogram
    plt.subplot(2,2,1)
    sns.histplot(x = column , data = df , color = 'skyblue' , edgecolor = 'black' , bins = 6)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    #KDE plot
    plt.subplot(2,2,2)
    sns.kdeplot(x = column , data = df , color = 'darkred')
    plt.title(f'KDE of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    #Boxplot
    plt.subplot(2,2,3)
    sns.boxplot(x = column , data = df , color = 'lightgreen')
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    plt.show()
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"For {column}:")
    print("Lower bound:", lower_bound)
    print("Upper bound:", upper_bound)
    print("Outliers count:", df[(df[column] < lower_bound) | (df[column] > upper_bound)].shape[0])
    print("-" * 50)

    # Return cleaned data
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def plot_categorical_analysis(df, columns, top_n=10):
    """
    Perform complete univariate analysis for categorical columns:
    - Displays summary stats (unique values, counts, mode)
    - Plots Count Plot, Pie Chart, and Bar Plot (Top N categories)
    """

    for column in columns:
        print("\n" + "*" * 20, f"{column}", "*" * 20)
        
        # Clean column (convert to string and drop blanks)
        df[column] = df[column].astype(str).str.strip()
        df_clean = df[df[column] != ""]

        # Summary
        print(f"\n Unique Values ({min(5, df_clean[column].nunique())} samples):")
        print(df_clean[column].unique()[:5])
        print(f"\n Number of Unique Values: {df_clean[column].nunique()}")
        print(f"\n Value Counts (Top {top_n}):\n{df_clean[column].value_counts().head(top_n)}")
        print(f"\n Mode: {df_clean[column].mode()[0]}")  #  FIXED HERE
        print("-" * 60)

        # Prepare Top N data
        top_values = df_clean[column].value_counts().nlargest(top_n)
        
        # --- Visualization ---
        plt.figure(figsize=(15, 12))

        # 1 Count Plot
        plt.subplot(3, 1, 1)
        sns.countplot(data=df_clean, x=column, order=top_values.index, palette='rocket')
        plt.title(f"Count Plot of Top {top_n} '{column}' Categories")
        plt.xlabel(column)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.tight_layout(pad=2)

        # 2 Pie Chart
        plt.subplot(3, 1, 2)
        top_values.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('rocket'))
        plt.title(f"Pie Chart of Top {top_n} '{column}' Categories")
        plt.ylabel("")  # Hide label

        # 3 Bar Plot
        plt.subplot(3, 1, 3)
        top_values.plot(kind='bar', color='coral', edgecolor='black')
        plt.title(f"Bar Plot of Top {top_n} '{column}' Categories")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.xticks(rotation=45)

        plt.tight_layout(pad=2)
        plt.show()
```

```python
# Load your dataset
data = pd.read_csv("Lol.csv")

# Select categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()

# Run analysis
plot_categorical_analysis(data, categorical_cols, top_n=10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datase
data = pd.read_csv("lol.csv")

# Identify categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()

# ----------- CONFIG ------------
TOP_N = 8  # Change this to 5 or 8 as per need
# --------------------------------

# Step 1: Summarize Top Categories
print(" Top Categories (% Share):\n")
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(data[col].value_counts(normalize=True).head(TOP_N) * 100)
    print("-" * 60)

# Step 2: Group Rare Categories (Keep only Top N)
def group_top_n(df, col, top_n=TOP_N):
    top_values = df[col].value_counts().nlargest(top_n).index
    df[col] = df[col].apply(lambda x: x if x in top_values else 'Other')
    return df

for col in categorical_cols:
    data = group_top_n(data, col)

# Step 3: Plot Univariate Analysis (Top N Categories)
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    order = data[col].value_counts().index  # Keep bars in order of frequency
    sns.countplot(data=data, x=col, order=order, palette='pastel', edgecolor='black')
    plt.title(f"Top {TOP_N} Categories in {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Step 4: Final Cleaned Summary
print("\n Cleaned & Grouped Categorical Data Summary:")
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(data[col].value_counts())
    print("-" * 50)
```
